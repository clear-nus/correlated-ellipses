````python
def render_images(np_x, n_rows=10, n_cols=10, imsize=64, channels=1, figsize=[4, 4]):
    plt.figure(figsize=figsize)
    np_x = np_x.reshape((n_rows,n_cols,imsize,imsize,channels))
    np_x = np.concatenate(np.split(np_x,n_rows,axis=0),axis=3)
    np_x = np.concatenate(np.split(np_x,n_cols,axis=1),axis=2)
    x_img = np.squeeze(np_x)
    plt.figure(figsize=figsize)
    plt.imshow(x_img, cmap='Greys_r')
    plt.xticks(np.arange(0, 64 * n_cols, 64))
    plt.yticks(np.arange(0, 64 * n_rows, 64))
    plt.grid()
    plt.show()
````

### Loading from file

````python
data = np.load('correlated_ellipses.npz')['dataset']
render_images(data[:100])
````

### Generating new samples

````python
data = synthetic.sample()
render_images(data)
````

### Conditional sampling
Fixes one factor in the batch

````python
data = synthetic.conditional_sample(0) # Fixed index 0, i.e., x position
render_images(data)
````
